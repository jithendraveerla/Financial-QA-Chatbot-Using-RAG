```
# embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

from transformers import pipeline

qa_model = pipeline(
    "question-answering",
    model="deepset/roberta-base-squad2"
)
```

```
def extract_text_from_pdf(pdf_file):

    doc = fitz.open(pdf_file.name)

    text = ""

    for page in doc:
        text += page.get_text()

    return text
```

```
def chunk_text(text, chunk_size=1000, overlap=200):

    chunks = []

    start = 0

    while start < len(text):

        end = start + chunk_size

        chunk = text[start:end]

        chunks.append(chunk)

        start += chunk_size - overlap

    return chunks
```

```
def process_pdf(pdf_file):

    try:

        # open PDF from filepath
        doc = fitz.open(pdf_file.name)

        text = ""

        for page in doc:
            text += page.get_text()

        chunks = chunk_text(text)

        vectors = []

        for i, chunk in enumerate(chunks):

            embedding = embedding_model.encode(chunk).tolist()

            vectors.append({
                "id": f"chunk-{i}",
                "values": embedding,
                "metadata": {"text": chunk}
            })

        index.upsert(vectors=vectors)

        return "Document processed successfully."

    except Exception as e:
        return f"Error: {str(e)}"
```

```
def ask_question(question):

    query_embedding = embedding_model.encode(question).tolist()

    results = index.query(
        vector=query_embedding,
        top_k=3,
        include_metadata=True
    )

    contexts = [m["metadata"]["text"] for m in results["matches"]]

    if len(contexts) == 0:
        return "The information is not available in the financial report."

    context = " ".join(contexts)

    try:
        result = qa_model(
            question=question,
            context=context
        )

        answer = result["answer"]

        if answer.strip() == "":
            return "The information is not available in the financial report."

        return answer

    except:
        return "The information is not available in the financial report."
```

```
with gr.Blocks() as demo:

    gr.Markdown("# Financial QA Bot")

    gr.Markdown("Upload a financial statement PDF and ask questions.")

    pdf_input = gr.File(label="Upload Financial Statement PDF")

    process_button = gr.Button("Process Document")

    status_output = gr.Textbox(label="Status")

    question = gr.Textbox(label="Question")

    answer = gr.Textbox(label="Answer", lines=8)

    ask_button = gr.Button("Submit")

    process_button.click(
        process_pdf,
        inputs=pdf_input,
        outputs=status_output
    )

    ask_button.click(
        ask_question,
        inputs=question,
        outputs=answer
    )

demo.launch(share=True)
```
